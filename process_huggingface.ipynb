# Process `thai_toxicity_tweet` for [huggingface/datasets](https://github.com/huggingface/datasets)

```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
from tqdm.auto import tqdm

def get_tweet(tweet_id):
    url = f'https://twitter.com/LllKnot/status/{str(tweet_id)}'
    user_agent_old_phone = 'Nokia5310XpressMusic_CMCC/2.0 (10.10) Profile/MIDP-2.1 '\
    'Configuration/CLDC-1.1 UCWEB/2.0 (Java; U; MIDP-2.0; en-US; '\
    'Nokia5310XpressMusic) U2/1.0.0 UCBrowser/9.5.0.449 U2/1.0.0 Mobile'
    headers = { 'User-Agent': user_agent_old_phone}
    try:
        with requests.get(url, headers=headers) as r:
            soup = BeautifulSoup(r.text,'lxml')
        return soup.find('div',class_='dir-ltr').text.strip()
    except:
        print(tweet_id)
        return 'TWEET_NOT_FOUND'
```

```python
df = pd.read_csv('annotated_corpus.txt',sep='\t',header=None)
df.columns = ['is_toxic','votes','tweet_id']
df['toxic_votes'] = df.votes.map(lambda x: x[1])
df['nontoxic_votes'] = df.votes.map(lambda x: x[3])
```

```python
tweet_texts = []
for tweet_id in tqdm(df.tweet_id.tolist()):
    tweet_texts.append(get_tweet(tweet_id))
```

```python
df['tweet_text'] = tweet_texts
df = df[['tweet_id','tweet_text','toxic_votes','nontoxic_votes','is_toxic']]
```
